Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees

```python
team_counts=df['Team'].value_counts() #Counting the no.of employees across each team
print(team_counts)
```

```python
total_employees = len(df) #Calculating the total no.of employees
print(total_employees)
```

```python
percentage=(team_counts / total_employees) * 100 #Calculating the percentage
print(percentage)
```

```python
distribution=pd.DataFrame({'Employee Count':team_counts, 'Percentage':percentage})
print(distribution) #Creating a new dataframe to show it in a understandable way
```

```python
plt.figure(figsize=(10, 6))
team_counts.plot(kind='bar', color='blue')
plt.title('Employee Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

This analysis reveals how employees are spread across different teams. Understanding this distribution helps in resource allocation and team management.

Identify the predominant age group among employees

```python
bins = [0, 20, 30, 40, 50, 60,70]  # Setting the age range
labels = ['0-20', '21-30', '31-40', '41-50', '51-60','61-100']#Setting the age group
#Creating a new column for age grouping
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
#Counting the newly created column
counts = df['Age Group'].value_counts()
#Checking the maximum emploee count in the group and the count too
predominant_age_group = counts.idxmax()
predominant_count = counts.max()
print(predominant_age_group)
print(predominant_count)

print("\nPredominant Value")
print(f"Your highest value is: {predominant_age_group} with {predominant_count} Employees")
```

```python
plt.figure(figsize=(10, 6))
counts.plot(kind='bar', color='green')
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Identifying the predominant age group provides insights into the workforce demographics. This can guide decisions related to training, benefits, and workplace culture.

Discover which team and position have the highest salary expenditure

```python
#Grouping teo columns and finding the sum salary
salary = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
#Finding the maximum salary
max = salary.loc[salary['Salary'].idxmax()]

print("Team and Positon with Highest Salary Expenditure:")
print(f"\n{max}")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=salary, x='Salary', y='Team', hue='Position', dodge=True)
plt.title('Salary Expenditure by Team and Position')
plt.xlabel('Total Salary Expenditure')
plt.ylabel('Team')
plt.grid(axis='x')
plt.show()
```

This analysis of salary expenditure reveals which teams and positions are costing the most. This information is vital for budgeting and financial planning.

Investigate if there's any correlation between age and salary, and represent it visually

```python
#Calculate Correlation
correlation = df['Age'].corr(df['Salary'])
print(f"Correlation between Age and Salary: {correlation}")
```

```python
#Visual representaion of correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Salary', color='blue')

# Added a regression line
sns.regplot(data=df, x='Age', y='Salary', scatter=False, color='red')

# Adding titles and labels
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)

# Shows the plot
plt.show()
```

Investigating the correlation between age and salary shows whether older employees tend to have higher salaries. A positive correlation may suggest that experience is rewarded with higher pay, while a weak or negative correlation could indicate different pay structures.
